```python
# Let's find VJ a new iphone!
import requests
import re
from bs4 import BeautifulSoup
```

```python
# get data
data = requests.get("https://slickdeals.net/deals/phone/")

# put into BeautifulSoup
soup = BeautifulSoup(data.text, "html.parser")
```

```python
# make sure it worked, if you see None, it didn't work.
# soup.text
```

```python
front_page_items = soup.find_all('div', {"class":"fpItem"})

# regular expression to get price
# reference: https://stackoverflow.com/q/46163913
price_re = re.compile(r"\$[,\d]+.?\d*")

# slickdeal's website only used https://slickdeals.net/f/...html
# so at each loop we need to append to the front
base_link = "https://slickdeals.net/"

# list to hold list of slick items ['title', 'price', 'link']
slick_items = []

for item in front_page_items:
    # class="itemTitle"
    title = item.find('a', {"class":"itemTitle"}).text
    
    # using regular expression to extract price
    match = price_re.search(title)
    if match:
        price = match.group()
        print(price.ljust(10), title[:100])
        
        link = base_link + item.find('a', href=True)['href']
        print(link + '\n')
    
        item = [title, price, link]
        slick_items.append(item)
```

```python
import pandas as pd

# we need additional clean up for the price
# where there are 2 numbers $200-$300 for example
header = ['title', 'price', 'link']

slick_df = pd.DataFrame(slick_items, columns=header)

# review dataframe
slick_df
```
