```python
# Inspired by: Python Web Scraping with Beautiful Soup and Regex
# https://www.youtube.com/watch?v=F1kZ39SvuGE

# Practice web scraping using BS4
# Website: Yahoo! Fiance Stock Page

import requests
from bs4 import BeautifulSoup
```

```python
# get the data
data = requests.get("https://finance.yahoo.com/gainers")

# load data into bs4
soup = BeautifulSoup(data.text, "html.parser")
```

```python
# Looping all the TR element
stock_list = []

# Looping through all the TR element tag
for tr in soup.find_all('tr'):
    # List comphension
    values = [td.text for td in tr.find_all('td')][:-2]
    
    # We do not want blank list (first record)
    if values:
        stock_list.append(values)
        
# Check out an example
print(stock_list[0])
```

```python
import pandas as pd

# Create the dataframe using web scrapped data
headers = ['Symbol', 'Name', 'Price', 'Change', '% Change', 'Volume',
           'Avg Vol (3 month)', 'Market Cap']

stock_df = pd.DataFrame(stock_list, columns = headers)

# Sample the dataframe
stock_df.head(10)
```
